```python
#                                  Importing Libraries
```

```python
import cv2
from cv2_plt_imshow import cv2_plt_imshow
import matplotlib.pyplot as plt
```

```python
#                               Importing Image
```

```python
img = cv2.imread("pencil.JPG")
cv2_plt_imshow(img)
plt.axis(False)
plt.show()
```

```python
#                     Convertion into RGB format
```

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_plt_imshow(img2)
plt.axis(False)
plt.show()
```

```python
#                               Conversion of Image into Grey Scale Image
```

```python
grey_img = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
cv2_plt_imshow(grey_img)
plt.axis(False)
plt.show()
```

```python
#                                  Inverting The Grey Scale Image
```

```python
inverted_img = cv2.bitwise_not(grey_img)
cv2_plt_imshow(inverted_img)
plt.axis(False)
plt.show()
```

```python
#                                      Blurry Image
```

```python
img_blur = cv2.GaussianBlur(inverted_img, (21,21), sigmaX=0,sigmaY=0)
cv2_plt_imshow(img_blur)
plt.axis(False)
plt.show()
```

```python
#                                        Pencil Sketch
```

```python
def pencil_sketch(x, y):
    return cv2.divide(x, 255-y, scale=256)
```

```python
pencil_sketch = pencil_sketch(grey_img,img_blur)
cv2_plt_imshow(pencil_sketch)
plt.axis(False)
plt.show()
```

```python
#                               Saving Image
```

```python
cv2.imwrite("pencil_sketch.jpeg", pencil_sketch)
```

```python
#                        Showing both Original Image and Pencil Sketch Image
```

```python
cv2_plt_imshow(img)
plt.axis(False)
plt.show()
```
